# Maximum Subarray
Given a integer array nums, find the contiguous subarray (containing at least one number) which has the largest sum and return that sum

## Example
```
input = [-2,1,-3,4,-1,2,1,-5,4]
output=6
cause subarray [4,-1,2,1] has the max sum =6
```

# Approach
- Note that the array values can be frm negative to positive
- Subarrays: [-2] is a sub array of the e.g. So is [-2,1], [-2,1,-3], [-2,1,-3,4]...

Initial approach
- Initially you'd consider looping through each subarray till you find the max
```
for(i=0...n-1) //For the start index
    for(j=i...j=n=1) //For the end index of the array starting at i

        //Now to compute the sum we'd need another loop
        for(k=i...k=j)
            //find sum
```
- This would then be O(n^3)

Better approach
- Say you've computed the subarray [-2,1,-3,4]
- To compute the result for the next we'd just need to add the next element -1
- The compare it to the initial sum

```
for(i=0...n-1)
    for(j=i...n-1)
        current sum + 1
```

**Even Better Approach**
- Disregard the negative prefixes

```
input = [-2,1,-3,4,-1,2,1,-5,4]

Algorithm:
MaxSum: [-2]
CuurentSum= -2
then add [1] but the **prefix** [-2] is negative, so we disregard it

MaxSum: [1]
MaxSubarray: [1]
CurrentSum= -1
then add [-3], the **prefix** [1] is positive, so keep it

MaxSum: [1]
MaxSubarray: [1]
CurrentSum= -2
then add [4], the **prefix** [-3] is negative, so disregard it and values before it

MaxSum: [4]
MaxSubarray: [4]
CurrentSum= 4
then add [-1], the **prefix** [4] is positive, so leave it be

MaxSum: [4]
MaxSubarray: [4]
CurrentSum= 3
then add [2], **BUT** even thou the prefix is negative, its prefix which forms the max sum is positive so leav it be

MaxSum: [5]
MaxSubarray: [4,-1,2]
CurrentSum= 5
then add [1]

MaxSum: [6]
MaxSubarray: [4,-1,2,1]
CurrentSum= 6
then add [-5]

MaxSum: [6]
MaxSubarray: [4,-1,2,1]
CurrentSum= 1
then add [4]

MaxSum: [6]
MaxSubarray: [4,-1,2,1]
CurrentSum= 5

Therefor

MaxSum: [6]
MaxSubarray: [4,-1,2,1]
```

This mimics a sliding window since we keep incrementing our right pointer, but the left pointer gets shifted only if the prefix is negative.

~ O(n)




````python
class EvenBetterApproach:
    def MaxSubarray(arr:list[int]):
        maxSubarray= arr[0] # INITIALLY the 1st index is the maxsubarray
        currentSum = 0

        for number in arr:
            if currentSum<0: # Means the prefix is negative
                currentSum=0
            currentSum += number

            maxSubarray = max(maxSubarray, currentSum)
        return maxSubarray

input = [-2,1,-3,4,-1,2,1,-5,4]
s = EvenBetterApproach

print(s.MaxSubarray(arr=input))
````

# Subarrays:
These are continuous chunks of an array

suppose we have:
```
arr = [1,2,3,4]
```

Examples of subarrays:
```
[1]
[2]
```
