# TwoSum

Given an input array OF integers, return incdices of the two numbers such that they add up to a specified target.

You may assume that an input has exactly one solution and you cant use the same element twice.

````python
class MySolution:
    def twoSum(self, arr, n:int):
        if len(arr)==0: return [-1, -1]

        hashMap = {}

        for i in range(len(arr)):
            if (n-arr[i]) in hashMap: #Cheks if the key is in the hasmap
                return [hashMap[n-arr[i]], i] #returns the index and the value (index in hashMap)
            hashMap[arr[i]]=i
        
class NeetSolution:
    def twoSum(self, nums:list[int], tar:int):
        if len(nums)==0: return [-1, -1]

        prevMap = {}

        for i, n in enumerate(nums):
            diff = tar - n
            if diff in prevMap:
                return [prevMap[diff], i]
            prevMap[n]=i
        return [-1, -1]
        
test = MySolution
a = []
n = 11
print(test.twoSum(test, a, n))
````
