# Anagrams

Given 2 str s and t, return true if they are anagrams and false if not

Think of:

- Size Equality
- Occurance Equality
    - A character with n counts, in s, should exist and also have n counts in t

Approaches:

    - Store each character as a key and the count as the value e.g. {a:3}

# 1. HashMap Approach:

## 1. Building the HashMaps

1. Your size is your base case
2. Create the empty HashMaps (dicts)
3. DON'T iterate through EACH string. Iterate through BOTH same time to lower the time complexity (utilising the equal length)
4. Find the characters of each string at each index
5. Assign an increment of 1 and the value associated the key (Character) to the current HashMap

## 2. Comparison
1. Iterate through 1 string
2. Compare the HashMaps at that key
3. If the amount of values aren't equal then return false

```python
class Solution1:
    def isAnagram(self, s, t) -> bool:
        if len(s) != len(t):
            return False
        
        sChars, tChars = {}, {}

        for i in range(len(s)):              # We usig the len intead of "i in s and i in t" to optimise the time complexity by iterating thru both simultaneously
            char_in_s = s[i]
            sChars[char_in_s] = 1 + sChars.get(char_in_s, 0)

            char_in_t = t[i]
            tChars[char_in_t] = 1 + tChars.get(char_in_t, 0) #gets the char or if the character doesnt exist yet it returns 0

        #=====================COMPARISON====================================================================================

        for key in sChars:
            if sChars[key] != tChars[key]: return False
        
        return True
        
```

# 2. Sorting

- if you can sort an anagram then the strings should be the same thing
- You can run an O(1) comparison to check if they are anagrams

Advantages
- Reduces space complexity since you dont have to store both HashMaps
- Faster comparison
- Interviewers dont counting complexity for sorting algorithms

Disadvantage
- Some sorting algorithms like bubble sort can take up to O(n^2) even though most are O(nlogn)
