## Series
Since a series is 1D we can use lists
We can have an array and convert it to a panda series like this:

### Converting to a series

```python
import pandas as pd
Sdata = [102, 300, 257, 30]

series = pd.Series(Sdata) #the data is the arg
```

Printing the series object creates a series with all our data arranged in a single column, with indices on the left and meta data at the bottom of the data-type and bits

```python
print(series)
```

### Index Labels

We can create our own index labels:

```python
series = pd.Series(Sdata, index=['a', 'b', 'c', 'd']) #the data is the arg, the index is a kwarg for the index labels
print(series)
```

The kwag could even be a tuple, a dictionay, a numpy array etc

### Accessing Values

We do this by utilising the loc propety
-loc means: Location by label

```python
print(f"Value accessed with loc['a'] = \033[1m{series.loc['a']}\033[0m") #loc is used to access the index labels, in this case 'a' which is 102
print(f"Value accessed with iloc[1] = \033[1m{series.iloc[1]}\033[0m") #iloc is used to access the index positions, in this case1 which is 300

#ANSI escape code for bold which is \033[1m and to reset it back to normal you can use \033[0m
```

### Replacing a value
You replace values normally by accessing the value at that index and assigning it a new value

```python
#Replacing
print("series before:")
print(series)

series.loc['a'] = 111
series.iloc[1] = 999

print()
print("series after replacement:")
print(series)
```

### Formatting
We can filter a list to get specific values we want by using a subscript operator [ ]


    -format series[series condition]
    -this is called boolean indexing

```python
print(series[series > 250]) 
#boolean indexing, this will return a series with only the values that are greater than 250, in this case 999 and 257
```

### Dictionaries

Dictionaries typically share the same operations with array with only a few differences

```python
#We'll use a dictionary of calories
calories_dict = {
    'day1': 420,
    'day2': 380,
    'day3': 390
}

calories_series = pd.Series(calories_dict)
print(calories_series)
```

Notice how the indices have already been replaced by the keys

Replacing remains the same too

```python
calories_series["day2"] = 380
print(calories_series)
```

#### Inserting a new Key-Value pair

```python
calories_series["day4"] = 450
print(calories_series)
```

#### Incrementing
We can also access and increment values

```python
calories_series["day4"] += 50
print(calories_series)
```

#### Filtering
This also remains the same

```python
filtered_calories = calories_series[calories_series > 385]
print(f"Days I OVER calorized:\n{', '.join(filtered_calories.keys())}.") #this will print the index labels of the filtered series, in this case 'day1', 'day3', and 'day4' since their values are greater than 385
#', ' -> separator for join method
#join method -> combines elements of an iterable into a single string based on the separator
```

## DataFrame

### Index Labels

```python
import pandas as pd
Ddata = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Course': ['Math', 'Science', 'History']
}

df = pd.DataFrame(Ddata, index=['student1', 'student2', 'student3']) #the data is the arg, the index is a kwarg for the index labels
print(df)
```

### Adding a new column

## Aggregation

Aggregate functions are functions that reduce a set of values to a single summary value

```python
import pandas as pd
df_by_name = pd.read_csv("data.csv", index_col="Name") #this will read the csv file and set the 'Name' column as the index of the dataframe
```

### Mean

#### All numeric columns

To find the mean to all numeric columns, do the following:

    - Call the mean aggregate method .mean()
    - Pass in the kwarg numeric_only=True
    - This ensures the method looks at the numeric values only

```python
print(df_by_name.mean(numeric_only=True)) #this will return a series with the mean of each numeric column in the dataframe
```

#### Single columns

The only difference here is that you're using the subscript operator to get your preferred column data

```python
print(df_by_name["Height"].mean()) #this will return the mean of the 'Height' column in the dataframe, which is a single value
```

### Sum

```python
print(df_by_name.sum(numeric_only=True)) #this will return a series with the sum of each numeric column in the dataframe
```

#### Single columns

## Data Cleaning
Fixing or Removing incomplete, incorrect or irrelevant data

    ~ 75% of work done with Pandas is for data cleaning

```python
import pandas as pd
df=pd.read_csv("data.csv") #this will read the csv file and set the 'Name' column as the index of the dataframe
```

### Dropping

Here's how you "drop" an irrelevant column

-Use the .drop() function
-Pass in the kwarg columns with the value being a list of the columns you want to "drop"
-Now assign that to a variable

```python
df_drop=df.drop(columns=["Type2", "No"]) #this will drop the 'Type2' column from the dataframe, inplace=True means that it will modify the original dataframe instead of returning a new one
print(df_drop)
```

    -This returns a new data frame intead of modifying the old one
    -To modify and existing data frame you'd use the kwarg inplace=True

### Handling Missing Data

    -Sometimes there's data that doesn't have required values from another column
    -e.g. Some Pokemons have type1 but no type2 and type2 might be required
    -We handle this by dropping that row or replacing the data in the column with something


#### Dropping Not Availables
    -We do that using df.dropna
    -Meaning "Drop Not Available"
    -You then pass in a kwarg of subset and assign it a list of columns
    -Meaning we'll drop that row if values of the specified column are naN
    -Don't forget to assign that to a variable

```python
df_both_types = df.dropna(subset=["Type2"]) #this will drop the rows where the 'Type2' column is null, which means that it will return a dataframe with only the rows where the Pokemon has a second type
print(df_both_types.to_string())
```

#### Replacing Not availables
    -We do that by using df.fillna
    -Meaning "Fill Not Available"
    -We pass in a dictionary
    -The key to our dict will be the column title
    -The value then it what you want to replace the missing values by

```python
df_replaced = df.fillna({"Type2": "None"}) #this will replace the null values in the 'Type2' column with the string "None", which means that it will return a dataframe with the null values replaced
print(df_replaced.to_string())
```

### Fixing Inconsistent Values

    -We can replace any instance of a value with another one

Let's replace any instance ot Grass types with GRASS

    -Access the column you're dealing with
    -Assign the column of replaced instances to it
    -The column of replaced instances consists of .replace() being used on the column being dealt with
    -Pass in a dictionary
    -The key is what you wanna replace
    -The value is what you wanna replace it by

```python
df["Type1"] = df["Type1"].replace({"Grass": "GRASS"}) #this will replace the null values in the 'Type2' column with the string "None", which means that it will return a dataframe with the null values replaced, this is another way to replace null values in a specific column
print(df)
```

    -You can pass in multiple K-V pairs

```python
df["Type1"] = df["Type1"].replace({"Grass": "GRASS",
                                    "Water": "H20"}) #this will replace the null values in the 'Type2' column with the string "None", which means that it will return a dataframe with the null values replaced, this is another way to replace null values in a specific column
print(df.to_string())
```

### Standardizing Text
When taking user input it is goo practise to keep the case of the strings consistent

    -Select the series (column)
    -Assign the standardized series to it

#### Strings

```python
df["Name"] = df["Name"].str.lower() #this will convert the 'Name' column to lowercase, which is a string method that can be applied to a series
print(df)
```

### Fixing Data Types

    -Select the series (column)
    -Assign the series with the new type to it
    -Utilize the astype() method
    -Pass in the type of your choice

```python
df["Legendary"] = df["Legendary"].astype(bool) #this will convert the 'Legendary' column to boolean, which means that it will return a dataframe with the 'Legendary' column converted to boolean values
print(df)
```
